```python
print("------ MENU ------")

menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320 
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25 
}

orders = []



while True:
    choice = input("Do you want to order (yes/no): ").lower()

    if choice == "no":
        break

    order = {}
    while True:
        item = input("Enter item name (or 'done' to finish this order): ").lower()

        if item == "done":
            break

        if item not in menu:
            print("Item not on menu!")
            continue

        qty = int(input("Enter quantity: "))

        order[item] = qty

    orders.append(order)

processed_orders = []

for order in orders:
    valid_order = {}
    for item, qty in order.items():

        if item not in menu:
            continue

        if qty > inventory[item]:
            qty = inventory[item]

        if qty > 0:
            valid_order[item] = qty

    processed_orders.append(valid_order)

print("\nProcessed Orders:")
print(processed_orders)

total = sum(qty * menu[item] for item, qty in order.items())
order_totals.append(total)

print("\nOrder Totals:")
print(order_totals)


items_sold = {}

for order in processed_orders:
    for item, qty in order.items():
        inventory[item] -= qty
        items_sold[item] = items_sold.get(item, 0) + qty

print("\nUpdated Inventory:")
print(inventory)

total_revenue = sum(order_totals)

print("\n---- DAY SUMMARY ----")
print("Total Revenue:", total_revenue)
print("Items Sold:", items_sold)


low_stock = [item for item, qty in inventory.items() if qty <= 5]

print("\nLow Stock Items (≤ 5 left):")
print(low_stock)


discounted_menu = {}

for item, price in menu.items():
    if inventory[item] > 30:
        discounted_menu[item] = price * 0.9
    else:
        discounted_menu[item] = price

print("\nMenu After Discount:")
print(discounted_menu)
```
