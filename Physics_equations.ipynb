```python
# Calculate Force Using Newton’s Second Law

mass = int(input("Enter mass in kilograms: "))
acceleration = int(input("Enter acceleration in meters per second squared: "))
force = mass * acceleration
print(f"{force}N")
```

```python
# Gravitational Potential Energy
mass = int(input("Enter mass in kilograms: "))
hieght = int(input("Enter mass in meters: "))
gravity = 9.8

Potential_Energy =  hieght * mass *gravity 
print(f"{Potential_Energy}J")
```

```python
# Kinetic Energy of an Object

mass = int(input("enter mass in kilograms:  "))
velocity = int(input("Enter velocity in meter per sec: "))

kinetic_energy = 1/2 * mass * velocity

print(f"{kinetic_energy}j")
```

```python
# Ohm’s Law –  Voltage Calculation
current = int(input ("enter the current in amperes: "))
resistane = int(input ("enter the resistance in ohm's: "))

Voltage = current * resistane
print(f"{Voltage}V")
```

```python
# Projectile Motion – Maximum Height

import math

Initial_velocity = float(input ("Enter the initial velocity in meters per sec: "))
angles  = float(input ("Enter the angle in degrees: "))
gravity = 9.8

theta = math.radians(angles)

height = (Initial_velocity**2 * (math.sin(theta)**2)) / gravity
print(height)
```

```python
# Harmonic Oscillator  – Period of a Spring


import math

# Input values
m = int(input("Enter mass (kg): "))
k = int(input("Enter spring constant k (N/m): "))

T = 2 * math.pi * math.sqrt(m / k)

print(f"{T:.2f} s")
```

```python
# Free Fall –  Time to Hit the Ground

import math

height = int(input("ennter height in meters: "))
gravity = 9.8

time = math.sqrt((2 * height) / gravity)
print(f"{time}sec")
```

```python
# Work Done by a Force
import math
Force = int(input("Enter force in newton : "))
displacement = int(input("Enter displaement in meters: "))
Angle = int(input("Enter angle in degrees: "))

theta_rad = math.radians(Angle)

Work = Force * displacement * math.cos(theta_rad)
print (f"{Work}j")
```

```python
# Power Consumption of a Device

work = int(input("enter work in joules: "))

time = int(input("enter time in sec: "))
power = time / work

print(f"{power}W")
```
